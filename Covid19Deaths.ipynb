```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
print('Modules are imported.')
```

<h2>Importing Data Sets</h2>

```python
corona_dataset_csv = pd.read_csv("Datasets/covid19_deaths_dataset.csv")
corona_dataset_csv.head(10)
```

```python
corona_dataset_csv.shape
```

<h2>Removing Unnecessary Columns</h2>

```python
corona_dataset_csv.drop(["Lat", "Long"],axis=1,inplace=True)
```

```python
corona_dataset_csv.head(10)
```

<h2>Aggregating Rows</h2>

```python
corona_dataset_aggregated = corona_dataset_csv.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head(10)
```

<h2>Visualising Deaths Among Various Countries <\h2>

```python
corona_dataset_aggregated.loc["China"].plot()
```

<h2>Calculating Good Measure</h2>

```python
corona_dataset_aggregated.loc["China"].diff().plot()
```

```python
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
countries = list(corona_dataset_aggregated.index)
maximum_death_rate=[]
for c in countries:
    maximum_death_rate.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["maximum_death_rate"] = maximum_death_rate    
```

```python
corona_dataset_aggregated.head()
```

<h2>Creating a new data frame with the requiredd column only<\h2>

```python
corona_data= pd.DataFrame(corona_dataset_aggregated["maximum_death_rate"])
```

```python
corona_data.head()
```

<h2>Preparing Happiness Data Set<\h2>

```python
happiness_report_csv = pd.read_csv("Datasets/worldwide_happiness_report.csv")
```

```python
happiness_report_csv.head()
```

```python
happiness_report_csv.drop(["Overall rank","Score", "Generosity", "Perceptions of corruption"],axis=1,inplace=True)
```

```python
happiness_report_csv.head()
```

```python
happiness_report_csv.set_index("Country or region", inplace=True)
```

```python
happiness_report_csv.head()
```

<h2>Joining both data sets<\h2>

```python
data = corona_data.join(happiness_report_csv, how="inner")
data.head()
```

<h2>Corealation Matrix<\h2>

```python
data.corr()
```
